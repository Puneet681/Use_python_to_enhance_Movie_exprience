```python
import cv2 
img = cv2.imread("D:\\backup\\IMG_20200814_120650652.jpg",1)
img.shape
w = (img.shape)[0]
h = (img.shape)[1]
print(img.shape)
rgb_list= []
for i in range (0,h-1):
    B,G,R=img[0,i]
    rgb_list.append((B,G,R))
for i in range (0,w-1):
    B,G,R=img[i,h-1]
    rgb_list.append((B,G,R))
for i in range (h-1,0,-1):
    B,G,R=img[w-1,i]
    rgb_list.append((B,G,R))
for i in range (w-1,0,-1):
    B,G,R=img[i,0]
    rgb_list.append((B,G,R))

for i in rgb_list:
    print(i)
len(rgb_list)
```
